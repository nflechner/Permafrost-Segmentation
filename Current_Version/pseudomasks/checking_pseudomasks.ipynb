```python
import os 

img_dir = '/Volumes/USB/Pseudomasks/orig_SNIC_pseudomasks/TRAIN/images'
imgs = os.listdir(img_dir)
filenames = [filename[:-4] for filename in imgs]

import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Randomly choose a filename from the list
# filename = random.choice(filenames)
# filename = '758_66_7550_2010_crop_5'
filename = '758_66_5025_2010_crop_114'

# Construct image and mask paths
img_path = f'/Volumes/USB/Pseudomasks/orig_SNIC_pseudomasks/TRAIN/images/{filename}.jpg'
mask_path = f'/Volumes/USB/Pseudomasks/orig_SNIC_pseudomasks/TRAIN/masks/{filename}.png'

# Load and plot the image and mask side by side
img = mpimg.imread(img_path)
mask = mpimg.imread(mask_path)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(img)
ax[0].set_title('Image (RGB)')
ax[0].axis('off')

ax[1].imshow(mask, cmap='gray')
ax[1].set_title('Mask (Binary)')
ax[1].axis('off')

plt.show()
```
