```python
import rasterio
import json
import os

def save_metadata_and_convert(directory, output_directory):
    # For each TIF file in the directory
    filenames = [f for f in os.listdir(directory) if not f.endswith('_aug.tif') and not f.startswith('._')]
    filenames =  ['761_71_5025_2018_crop_50', '761_71_5050_2018_crop_95', '764_74_7525_2018_crop_48', '760_75_0000_2015_crop_132', '761_72_0050_2018_negcrop_30', '755_70_7500_2010_crop_15', '764_74_5075_2018_crop_77', '763_73_5000_2018_negcrop_8', '763_72_7500_2018_crop_20', '759_76_5025_2013_negcrop_0', '764_74_2575_2018_crop_106', '749_61_0000_2019_crop_88', '760_75_0000_2015_negcrop_5', '760_71_0075_2018_negcrop_4', '761_73_0025_2018_crop_37', '763_73_5025_2018_crop_136', '762_76_7575_2015_negcrop_0', '759_73_0075_2013_crop_85', '761_73_2550_2018_negcrop_37', '758_74_5050_2013_crop_124', '758_70_0050_2013_crop_53', '758_73_5025_2013_negcrop_0', '756_70_5000_2010_crop_38', '747_68_2500_2014_negcrop_0', '761_72_0075_2018_negcrop_54', '761_76_7550_2015_crop_14', '760_71_0075_2018_negcrop_15', '760_74_7525_2018_negcrop_47', '763_73_5075_2018_negcrop_1', '755_70_7550_2010_negcrop_0', '760_74_7525_2018_negcrop_40', '761_73_0075_2018_negcrop_2', '761_76_7525_2015_negcrop_1', '746_57_5025_2015_negcrop_3']
    failed_count = 0  # Counter for failed instances
    for i, tif_file in enumerate(filenames):
        tif_file = tif_file + '.tif'
        # Read TIF with rasterio
        filepath = os.path.join(directory, tif_file)
        try:
            with rasterio.open(filepath) as src:
                # Extract metadata
                metadata = {
                    'transform': src.transform.to_gdal(),
                    'crs': src.crs.to_string(),
                    'bounds': src.bounds._asdict()
                }
                
                # Save metadata to JSON
                json_path = os.path.join(output_directory, f"{tif_file.split('.')[0]}_metadata.json")
                with open(json_path, 'w') as f:
                    json.dump(metadata, f)
        except rasterio.errors.RasterioIOError as e:
            print(f"Failed to open {filepath}: {e}")
            failed_count += 1

        try:
            with rasterio.open(filepath) as src:
                # Extract metadata
                metadata = {
                    'transform': src.transform.to_gdal(),
                    'crs': src.crs.to_string(),
                    'bounds': src.bounds._asdict()
                }
                
                # Save metadata to JSON
                json_path = os.path.join(output_directory, f"{tif_file.split('.')[0]}_metadata.json")
                with open(json_path, 'w') as f:
                    json.dump(metadata, f)
        except rasterio.errors.RasterioIOError as e:
            print(f"Failed to open {filepath}: {e}")
            failed_count += 1
        
        print(f"{i} files done")

    print(f"Total failed instances: {failed_count}")

input_dir = "/Volumes/USB/Orig_200m_TIFS/rgb"
output_dir = "/Volumes/USB/Orig_200m_TIFS/tif_metadata"
save_metadata_and_convert(input_dir, output_dir)
```
