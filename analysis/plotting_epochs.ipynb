```python
import pandas as pd

path = "/Users/nadja/Documents/1. Project/Thesis/Data/analysis/performance_metrics.csv"
data = pd.read_csv(path)
```

```python
data.head()
```

```python
data = data[[data.columns[3], data.columns[0], data.columns[2], data.columns[1]]]
```

```python
import matplotlib.pyplot as plt 

labels = ['Validation mIoU (background)','Validation mIoU (target)','Validation Loss', 'Training Loss']

plt.figure(figsize=(10, 6))
for i, column in enumerate(data.columns):
    plt.plot(data.index, data[column], label=labels[i])
plt.title('Performance Metrics Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Metrics')
plt.legend(loc='upper right', bbox_to_anchor=(1, 1 - (1/10)))
plt.show()
```
